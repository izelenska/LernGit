Удерживать пользователей: Пользователи, которые начинают использовать продукт в определенный период времени и продолжают использовать продукт через определенный период времени, являются зарезервированными пользователями.

```python
# посчитаем доход для каждого клиента
df_5['revenue']=df_5['price']*df_5['order_item_id']
df_5_rev=df_5
df_5_rev
```

```python
df_5_rev.dtypes
```

```python
#Определим в OrderPeriod месяц и год когда была совершена покупка
df_5_rev['order_period'] = pd.to_datetime(df_5_rev['order_approved_at']).dt.strftime('%Y-%m')
df_5_rev
```

```python
df_5_rev.set_index('customer_id', inplace=True)
df_5_rev['cohort_group'] = df_5_rev.groupby(level=0)['order_approved_at'].min().dt.strftime('%Y-%m')
df_5_rev.reset_index(inplace=True)
df_5_rev.head()
```

```python
grouped = df_5_rev.groupby(['cohort_group', 'order_period'])
# количество уникальных пользователей и общее количество заказов и выручка за период
# count the unique users, orders, and total revenue per Group + Period
cohorts = grouped.agg({'customer_id': pd.Series.nunique,
                       'order_id': pd.Series.nunique,
                       'revenue': np.sum})
# переименуем некоторые столбцы для большей наглядности
# make the column names more meaningful
cohorts.rename(columns={'customer_id': 'total_customer_id',
                        'order_id': 'total_order_id'}, inplace=True)
cohorts.head(15)
```

```python
# cohort_period определим как порядковый номер месяца в массиве начинающийся с 1.
def cohort_period(df_5_rev):
    #cohorts['cohort_period'] = np.arange(len(cohorts)) + 1
    df_5_rev['cohort_period'] = np.arange(len(df_5_rev))+1 # отсчет с 1
    return df_5_rev
cohorts = cohorts.groupby(level=0).apply(cohort_period)
cohorts.head()
```
